### 2.1 Introducción
El electrocardiograma (ECG) es una representación gráfica de la actividad eléctrica del corazón. La señal de ECG es una señal bioeléctrica que se puede medir en la superficie de la piel. La señal de ECG es ampliamente utilizada en la práctica clínica para evaluar la actividad eléctrica del corazón y diagnosticar enfermedades cardíacas. La señal de ECG es una señal periódica que se puede dividir en varios segmentos, como el complejo QRS, la onda P y el segmento ST (Figura 1). ### 2.3 Detección de peaks

La detección de peaks en la señal de ECG es una tarea fundamental en el procesamiento de señales biomédicas. La detección de peaks permite identificar los complejos QRS, que son los peaks más prominentes en la señal de ECG. La detección de peaks es una tarea desafiante debido a la variabilidad de la señal de ECG y al ruido presente en la señal. Existen varios métodos para detectar peaks en la señal de ECG, como el algoritmo de Pan-Tompkins, el algoritmo de Hamilton-Tompkins, el algoritmo de Christov, entre otros. Para este ejercicio, utilizaremos una metodología simple basada en el umbral de la señal.

#### Algoritmo de detección de peaks

Este algoritmo identifica peaks en una señal discreta basada en tres criterios:
1.	Debe ser mayor que un umbral predefinido (threshold).
2.	Debe ser un máximo local, es decir, debe ser mayor que sus vecinos inmediato izquierdo y derecho.

El enfoque del algoritmo es recorrer la señal y aplicar estas condiciones en cada punto.

Paso a paso del algoritmo
1.	**Entrada:** Se recibe una señal en forma de lista o array (signal) y un valor umbral (threshold).
2.	**Inicialización:** Se crea una lista vacía `peaks` para almacenar los índices de los peaks detectados.
3.	**Recorrido de la señal:** Se itera sobre los elementos de la señal, comenzando en el segundo elemento (`i = 1`) y terminando en el penúltimo (`i = len(signal) - 2`). Esto se hace para evitar acceder a elementos fuera de los límites al comparar con vecinos.
4.	**Evaluación de cada punto en la señal:**
	* Si `signal[i]` es mayor que el umbral threshold.
	* Si `signal[i]` es mayor que el elemento anterior `signal[i - 1]`.
	* Si `signal[i]` es mayor que el elemento siguiente `signal[i + 1]`.
	* Si las tres condiciones se cumplen, el índice `i` se agrega a la lista `peaks`.
5.	Salida: Se devuelve la lista peaks convertida en un array de NumPy (`np.array(peaks)`).




````python
def detect_peaks(signal, threshold):
    peaks = []
    for i in range(1, len(signal) - 2):
        if signal[i] > threshold and signal[i] > signal[i - 1] and signal[i] > signal[i + 1]:
            peaks.append(i)
    return np.array(peaks)

# Calcular el umbral de la señal correspondiente al 50% del valor máximo
threshold = 0.5 * np.max(ecg_10)
print(f"El umbral es: {threshold}")

# Detectar peaks en la señal de ECG
peaks_10 = detect_peaks(ecg_10, threshold)
print(f"Se detectaron {len(peaks_10)} peaks en la señal de ECG")
print(f"Los peaks detectados estan en las posiciones: {peaks_10}")

````

La función anterior entrega como resultado los índices de los latidos detecteados. Estos deben ser convertidos a segundos para poder calcular la frecuencia cardiaca, mediante la divición por su frecuencia de muestreo. 

$$ \text{tiempo latidos}=\dfrac{\text{índices}}{\text{frecuencia de muestreo}}$$

Escribiendolo como codigo:

````python

tiempo_latidos = peaks_10/frecuencia_muestreo
print(f"Los latidos ocurren en los siguientes tiempos: {tiempo_latidos}")
````

Con los tiempos de los latidos, es posible incluirlos en el gráfico de la señal de ECG para visualizar los peaks detectados. A continuación se muestra un ejemplo de como graficar la señal de ECG con los peaks detectados.

````python
fig = plt.figure(figsize=(12, 6))
plt.plot(tiempo_10, ecg_10)  # Recive el tiempo y la señal
plt.plot(tiempo_10[peaks_10], ecg_10[peaks_10], 'ro')  # Recive el tiempo y la señal
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [mV]')
plt.show()
````
